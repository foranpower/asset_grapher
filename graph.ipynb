```python
#from data
last_time = df.time.iloc[-1]
last_time
df
```

```python
###user selection below 4
effect_start_offset  = datetime.timedelta(days=-7)
# effect_end_offset = cause_start_offset
# cause_start_offset = anchor
cause_end_offset = datetime.timedelta(days=7)
offset = datetime.timedelta(days=-7) #offset between events
central_days_offset_should_be_zero = datetime.timedelta(days=2) # should be zero --but positive for into future
central_hour = 14
central_minute = 53
```

```python
try:
    df= df[['time', 'Open']]
except:
    df= df[['time', 'open']]
    df.rename(columns = {'open':'Open'}, inplace = True)
```

```python
#df.set_index(['time'], inplace=True)
df
#print(df.to_string())
```

```python
###open is all negative in time so open0 is present.
df_merged = pd.DataFrame()
df_mask = pd.DataFrame()
for i in range(0, 12):
    central = last_time + (i * offset)
    start2 = central + effect_start_offset
    start = start2.replace(hour=0, minute=0)
    end2 = central + cause_end_offset
    end  = end2.replace(hour=23, minute=59)
    mask = (df['time'] > start) & (df['time'] <= end)
    df_mask = df.loc[mask]
    df_mask = df_mask.rename(columns={'Open': f"Open{-i}"})
    df_mask['time'] = df_mask['time'] -(i*offset)
    try:
        df_merged = pd.merge(df_mask, df_merged,  how='outer', on='time', sort = True)
    except:
        df_merged = df_merged if not df_merged.empty else df_mask
```

```python
#anchorloc = df_merged.Open1.last_valid_index()
#anchorloc = df_mask2.Open0.last_valid_index()
df_merged2 = copy.deepcopy(df_merged)
#df_merged2 = df_merged2.fillna(method='ffill')
for i in range(0, 12):
    central = last_time +central_days_offset_should_be_zero #+ (i * offset)
    central = central.replace(hour=central_hour, minute=central_minute)
    mask2 = (df_merged['time'] <= central)
    df_mask2 = df_merged.loc[mask2]
    anchorloc = df_mask2[[f'Open{-i}']].last_valid_index()
    anchor_value = df_merged2[[f'Open{-i}']].iloc[anchorloc]
    df_merged2[[f'Open{-i}']]= df_merged2[[f'Open{-i}']] - anchor_value
```

```python

time = df_merged2.index.get_level_values("time")
# Getting info for plotting the day/night indicator
# time[0].date() picks out 00:00 (midnight) and then we add 9.5 hours to get 9:30 am.
start_morning = pd.to_datetime(time[0].date()) + pd.Timedelta(
    hours=8.5
)
end_morning = pd.to_datetime(time[-1].date()) + pd.Timedelta(
    hours=30
)
num_mornings = (end_morning - start_morning).days
# Now we build up the morning times, every day at 6 am
mornings = [
    start_morning + timedelta(days=x) for x in range(num_mornings)
]
for morning in mornings:
    fig.add_vrect(
# Highlighted region starts at 930 am and ends at 4 pm, every day.
        x0=morning,
        x1=morning + timedelta(hours=6.5),
        fillcolor="brown",
        opacity=0.1,
        line_width=0,
    )
# fig.show()
fig.update_traces(connectgaps=True)
fig.write_html( 'output_file_name.html', auto_open=True )
```
